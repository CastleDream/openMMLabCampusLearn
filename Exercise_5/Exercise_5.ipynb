## 2.3 ControlNet

`````python
import cv2
import numpy as np
import mmcv
from mmengine import Config
from PIL import Image

from mmagic.registry import MODELS
from mmagic.utils import register_all_modules

import matplotlib.pyplot as plt

register_all_modules()
`````

### 2.3.1 从线稿生成

`````python
# 载入ControlNet模型
cfg = Config.fromfile('/content/mmagic/configs/controlnet/controlnet-canny.py')
controlnet = MODELS.build(cfg.model).cuda()
`````

`````python
# 输入Canny边缘图
control_url = 'https://user-images.githubusercontent.com/28132635/230288866-99603172-04cb-47b3-8adb-d1aa532d1d2c.jpg'
control_img = mmcv.imread(control_url)
control = cv2.Canny(control_img, 100, 200)
control = control[:, :, None]
control = np.concatenate([control] * 3, axis=2)
control = Image.fromarray(control)
`````

`````python
plt.imshow(control)
`````

`````python
prompt = 'Room with blue walls and a yellow ceiling.'

output_dict = controlnet.infer(prompt, control=control)
samples = output_dict['samples']
for idx, sample in enumerate(samples):
    sample.save(f'/content/drive/MyDrive/OpenMMLab/Exercise_5/sample_{idx}.png')
controls = output_dict['controls']
for idx, control in enumerate(controls):
    control.save(f'/content/drive/MyDrive/OpenMMLab/Exercise_5/control_{idx}.png')
`````

### 2.3.2 从pose生成

`````python
import torch
torch.cuda.empty_cache()
`````

`````python
cfg2 = Config.fromfile('/content/mmagic/configs/controlnet/controlnet-pose.py')
# convert ControlNet's weight from SD-v1.5 to Counterfeit-v2.5
cfg2.model.unet.from_pretrained = 'gsdf/Counterfeit-V2.5'
cfg2.model.vae.from_pretrained = 'gsdf/Counterfeit-V2.5'
cfg2.model.init_cfg['type'] = 'convert_from_unet'
controlnet2 = MODELS.build(cfg2.model).cuda()
# call init_weights manually to convert weight
controlnet2.init_weights()
`````

`````python
control_url = 'https://user-images.githubusercontent.com/28132635/230380893-2eae68af-d610-4f7f-aa68-c2f22c2abf7e.png'
control_img = mmcv.imread(control_url)
control = Image.fromarray(control_img)
plt.imshow(control)
`````

`````python
prompt = 'masterpiece, best quality, sky, red hair, skirt, sailor collar, looking at viewer, long hair, building, bangs, neckerchief, long sleeves, sunny sky, power lines, shirt, cityscape, pleated skirt, scenery, blunt bangs, city, daylight, black sailor collar,smile face, big eyes'
`````

`````python
output_dict2 = controlnet2.infer(prompt, control=control, width=512, height=512, guidance_scale=7.5)
samples2 = output_dict2['samples']
for idx, sample in enumerate(samples2):
    sample.save(f'/content/drive/MyDrive/OpenMMLab/Exercise_5/sample_{idx}_pose.png')
controls2 = output_dict2['controls']
for idx, control in enumerate(controls2):
    control.save(f'/content/drive/MyDrive/OpenMMLab/Exercise_5/control_{idx}_pose.png')
`````
